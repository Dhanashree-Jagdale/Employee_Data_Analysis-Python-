# Handling Missing Data 

```python
#Handling missing values in [AGE] column:
df['Age']=df['Age'].fillna(df['Age'].median())    #fills average age in missing places
df=df[df['Age']>0]                              #(>0) cleans unrealistic values
df['Age'].describe()                             #shows the column statistics
```

```python
#Handling missing values in [Salary] column:
df['Salary']=df['Salary'].fillna(df['Salary'].median())
df=df[df['Salary']>0]
df['Salary'].describe()
```

```python
df.head()
```

```python
#Handling values in [Gender] column:
#Standardizing gender into only two columns [Male] [Female]:
df['Gender'].unique()     #this finds unique values from the column
```

```python
#Converting all values to lowercase
df['Gender']=df['Gender'].str.lower().str.strip()
```

```python
df['Gender']=df['Gender'].replace({
    'm':'Male',
    'male':'Male',
    'f':'Female',
    'female':'Female'
})
df['Gender'].value_counts()
```

```python
df['Gender'].unique() 
```

```python
#Cleaning Experience Column:
#Converting Experience into a numerice column (years):
#Experience(int)
df['Experience'].unique()
```

```python
#Converting into lowercase
df['Experience']=df['Experience'].str.lower().str.strip()
```

```python
word_to_num={
    'one':'1',
    'two':'2',
    'three':'3',
    'four':'4',
    'five':'5'
}
df['Experience']=df['Experience'].replace(word_to_num,regex=True)
```

```python
#Extracting numeric Values & Converting to Integer
df['Experience']= df['Experience'].str.extract(r'(\d+)').astype(int)
df[['Experience']].head()
```

```python
#resetting index values 
df = df.reset_index(drop=True)          #(drop) don’t keep old index
```

```python
df[['Experience']].head()
```

```python
#Cleaning City Comun
df['City'].unique()
```

```python
#Removed leading & trailing spaces
df['City']=df['City'].str.strip()
```

```python
#Standardizing Capitalization
df['City']=df['City'].str.title()
```

```python
#Fixing differnt spellings
#Bangalore & Bengaluru should be one city
df['City']=df['City'].replace({'Bengaluru':'Bangalore'})
```

```python
#Verifying cleaned cities
df['City'].value_counts()
```

```python
#Cleaning the Join_Date Column:
df['Join_Date'].dtype          #checking type first
```

```python
#Checking missing value:
df['Join_Date'].isna().sum()   #isna() is used to check for actual missing values
```

```python
#Finding the invalid rows
bad_dates = df[df['Join_Date'].isna()]
bad_dates
```

```python
#Taking only the year from the date and storing it in a new column
df['Join_Year'] = df['Join_Date'].dt.year
```

```python
#Cleaning Department Column 
df['Department'].unique()
```

```python
#filling the missing values
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
```

```python
df.isna().sum()
```
